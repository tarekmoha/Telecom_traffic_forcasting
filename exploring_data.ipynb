```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df= pd.read_csv("full_grid.csv", parse_dates=["time"])
```

```python
df.head()
```

```python
df.isna().sum()
```

## Get Random Grid:

```python
df_g = df.groupby("grid_num")
grids = list(df_g.groups.keys())
random_g = np.random.choice(grids)
df_g = df_g.get_group(random_g)
```

```python
random_g
```

### Full 2 months data

```python
fig, ax = plt.subplots(figsize = (20, 5))
ax.plot(df_g["time"], df_g["rate"])
```

### Null Values:

```python
df_g.isna().sum()
```
